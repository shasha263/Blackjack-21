```python
import random
class Blackjack:
    def __init__(self,playercards,rand_num,dealercards,sm_dealer,sm_player):
        self.playercards=playercards
        self.dealercards=dealercards
        self.rand_num=rand_num   
        self.sm_dealer=sm_dealer
        self.sm_player=sm_player
        
    def card_info(self):        
        self.playercards=[]
        self.dealercards=[]
        self.rand_num=[]
        card_types=['Clubs', 'Diamonds', 'Hearts', 'Spades']        
        for n in card_types:
            self.rand_num.append(10)
            self.rand_num.append(10)
            self.rand_num.append(10)
            for i in range(1,11):
                self.rand_num.append(i)
        #print(self.rand_num) #testing purpose
    
    def random(self):           
        
        for i in range(1,3):
            self.playercards.append(random.choice(self.rand_num))    
        print(f"Player's two cards are: {self.playercards}")          
        for i in range(1,3):
            self.dealercards.append(random.choice(self.rand_num))  
        print(f"Dealer's one card is {self.dealercards[0]}")  

    def card_show(self):
        print(f"Player's cards are: {self.playercards}") 
        print(f"Dealer's cards are: {self.dealercards}")

    def score(self):    
        self.sm_dealer= sum(self.dealercards)    
        self.sm_player= sum(self.playercards) 
  
    def hit(self): 
             
        if self.sm_player > 21:
            print("Player Busted") 
        elif self.sm_player <= 21:
            self.playercards.append(random.choice(self.rand_num))  
            print(f"Player's cards are: {self.playercards}")                 
        else:
            print("Player Busted")         
    
    def stand(self):
        if self.sm_player > self.sm_dealer:
            print('Blackjack :-)')                                
        elif self.sm_dealer < 17 and self.sm_dealer <self.sm_player:
            self.dealercards.append(random.choice(self.rand_num))  
            print(f"Dealer's cards are: {self.dealercards}") 
        elif self.sm_dealer > self.sm_player and self.sm_dealer>=17:
            print('Player Busted!!!')                    
        else:
            print("Dealer Busted")              
           
    
```
